```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 50)                      # 49 elements
y = 2*x + np.random.normal(0, 5, len(x))  # also 49 elements

plt.plot(x, y)
plt.title("Trend Example")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 50)  # 49 values
y_noise = 2*x + 10*np.sin(0.3*x) + np.random.normal(0, 10, len(x))

plt.plot(x, y_noise)
plt.title("Trend + Seasonality + Noise")
plt.show()
```

```python
import pandas as pd
data = pd.Series(y_noise)
data.rolling(10).mean().plot(label='Rolling Mean') #It’s the average of the last n data points in a sliding window.Smooths out short-term fluctuations.Highlights long-term trends.
data.rolling(10).std().plot(label='Rolling Std')#It’s the standard deviation of values within the moving window, showing how variable the data is in that window.Detects volatility or stability changes in time series.
plt.legend()
plt.show()
```

```python
data_diff = data.diff().dropna()
plt.plot(data_diff)
plt.title("Differenced Series (Stationary)")
plt.show()
```
