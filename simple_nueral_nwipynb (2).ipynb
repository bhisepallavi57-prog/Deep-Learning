```python
import numpy as np

#simple dataset: input-->output
X = np.array([[0],[1], [2],[3]])
y = np.array([[0], [1], [2], [3]])

#initialise weights and biases
np.random.seed(42)
W1 = np.random.randn(1,2)
b1= np.zeros((1,2))
W2 = np.random.rand(2,1)
b2 = np.zeros((1,1))

#activation function
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

#forward propogation
#step 1:input-> hidden
z1 = np.dot(X, W1) + b1
a1 = sigmoid(z1)

#step2: hidden->output
z2 = np.dot(a1, W2) + b2
output = z2 # no activation function for regression

print("Predicted output:")
print(output)
```

```python
loss = np.mean((y - output) ** 2)
print("Loss:" , loss)
```

```python
import numpy as np

#input with 2 fatures
X= np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

#lets make the target = sum of inputs
# y = x1+x2
y = np.array([[0], [1], [1], [2]])

#initialise weights and biases
np.random.seed(42)
#2->2->1
W1 = np.random.randn(2, 2)
b1 = np.zeros((1,2))
W2 = np.random.randn(2, 1) #hidden->output
b2 = np.zeros((1,1))

#activation function
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

#forward propogation
#step1:input>hidden
z1= np.dot(X, W1) + b1
a1= sigmoid(z1)

#step2: hidden>output
z2 = np.dot(a1, W2) + b2
output = z2

print("Predicted output:")
print(output)

#compute loss(mse)
loss =  np.mean((y- output) ** 2)
print("Loss:", loss)
```
